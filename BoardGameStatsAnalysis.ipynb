```python
# import the libraries
import pandas as pd
import json
from urllib.request import urlopen

# url to the data stored on GitHub
url = "https://raw.githubusercontent.com/gz97/board-games-analysis/refs/heads/main/BGStatsExport.json"

# store the response of URL 
response = urlopen(url)


# storing the JSON response  
# from url in data 
data_json = json.loads(response.read()) 

# dictionary with ALL data
data_json["challenges"]
```

```python
# split the dictorionatry into separate tables, such as Challenges, Players, Games, etc
# get all the possible tables first
for k in data_json:
    print(k)

# tags
# groups
# players
# locations
# games
# plays
# challenges
# deletedObjects
# userInfo

tags_json = data_json["tags"]
groups_json = data_json["groups"]
players_json = data_json["players"]
locations_json = data_json["locations"]
games_json = data_json["games"]
plays_json = data_json["plays"]
challenges_json = data_json["challenges"]
deletedObjects_json = data_json["deletedObjects"]
userInfo_json = data_json["userInfo"]    
```

```python
# Transform user data into a df for analysis
userInfo_df = pd.json_normalize(userInfo_json)
```

```python
# Transform tags data into a df for analysis
tags_df = pd.json_normalize(tags_json)
tags_df
```

```python
# Transform groups data into a df for analysis

# change the metaData value into a real dictionary from a string and add a missing metaData value to any item in the list that doesn't have it
default_metadata = None

for item in groups_json:
    if "metaData" not in item:
        item["metaData"] = default_metadata  # Assign default metadata
    elif isinstance(item["metaData"], str):  # Convert string metadata to dictionary
        item["metaData"] = json.loads(item["metaData"])

#groups_json
groups_df = pd.json_normalize(groups_json, sep = "_") 
#groups_df = pd.DataFrame(groups_json)      
groups_df
```
